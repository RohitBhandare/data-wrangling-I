```python
# 5. to store data into a new dataframe based on a specific condition

# The copy() method is used to ensure that the new dataframe is a separate copy of the 
# original dataframe, rather than a reference to the original dataframe.

# Create a new dataframe with rows where the age is greater than or equal to 18

new_data = data[data['age'] >= 18].copy()

print(new_data)
```

```python
# 6. to add a new feature with calculations on a column in a Pandas dataframe

# Add a new feature named "total_score" that is the sum of the "G1", "G2", and "G3" columns
data['total_score'] = data.apply(lambda row: row['G1'] + row['G2'] + row['G3'], axis=1)

# The apply() method applies the lambda function to each row of the dataframe, 
# and the axis=1 parameter tells it to apply the function across each row instead of each column.

# Add a new feature named "total_score" that is the sum of the "G1", "G2", and "G3" columns
data = data.assign(total_score=lambda x: x['G1'] + x['G2'] + x['G3'])

# Add a new feature named "total_score" that is the sum of the "G1", "G2", and "G3" columns
data['total_score'] = data['G1'] + data['G2'] + data['G3']


data
```

```python
# 7. To delete a particular column from an existing Pandas dataframe

# Delete the "absences" column from the dataframe

data = data.drop('absences', axis=1)

# The drop() method takes the column name to be deleted as the first argument, and the axis=1 
# argument specifies that we want to delete a column (as opposed to a row).

data

# Delete the "absences" column from the dataframe
del data['absences']

# This code will remove the "absences" column from the dataframe in-place. 
# The del keyword is used to delete a variable, in this case the specified column in the dataframe.

# Note that when you use this method to delete a column, it modifies the original dataframe without creating a copy. 
# If you want to keep the original dataframe intact, you should make a copy of the dataframe before using del to delete a column.
```

```python
# define a function in a Pandas dataframe

# Define a function that takes a row of the dataframe as input and returns the sum of G1, G2, and G3 columns
# method-1
def total_grade(row):
    return row['G1'] + row['G2'] + row['G3']

# Apply the function to the dataframe using the apply() method
data['total_grade'] = data.apply(total_grade, axis=1)
data


# method-2
# Define a lambda function to calculate the total grade for each row
total_grade = lambda row: row['G1'] + row['G2'] + row['G3']

# Apply the lambda function to the dataframe using the apply() method
data['total_grade'] = data.apply(total_grade, axis=1)

```

```python
# To convert a numerical variable into a categorical variable in Pandas
# The cut() function divides a range of values into bins and labels the bins with the corresponding categories

# Define the bins and labels for the categorical variable
bins = [0, 50, 70, 100]
labels = ['low', 'medium', 'high']

# Use the cut() function to convert the "G1" column into a categorical variable
data['G1_category'] = pd.cut(data['G1'], bins=bins, labels=labels)

# The bins list defines the boundaries for the different categories, and the labels list specifies the label for each category. 
# We then use the cut() function to convert the "G1" column into a categorical variable called "G1_category".
data
```
